```python
# Function to delete values in strings 
def data_cleaning(x,y,z):
    '''Takes in dataframe x, column y,
    replaces unwanted values z, with empty space
    to remove unwanted characters'''
    x[y]=x[y].str.replace(z,'')
    return x
```

```python
data_cleaning(df_imdb,'Title','\xa0')
```

```python
# check to see if data_cleaning function worked
df_imdb['Title'][27]
```

```python
'''Merge IMDb and Box Office Mojo dataframes by common title.
    Inner merge using Box Office Mojo because it has fewer titles'''
df_joined=df_mojo.merge(df_imdb,how='inner',left_on='Title',right_on='Title')
df_joined
```

```python
# save joined dataframe to .csv file
df_joined.to_csv('joined_df.csv')
```

```python
df_joined.info()
```

```python
#removing unnecessary columns
df_joined=df_joined.drop(['Unnamed: 0_x','Unnamed: 0_y','Year_y'],axis=1)
df_joined
```

```python
# removing unnecessary column
df_joined=df_joined.drop('Color',axis=1)
df_joined
```

```python
# rename odd column header
df_joined=df_joined.rename({'Year_x':'Year'},axis=1)
df_joined
```
