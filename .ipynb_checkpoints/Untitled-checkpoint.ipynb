```python
import locale
import pandas as pd
locale.setlocale(locale.LC_ALL,'')

df['Worldwide Lifetime Gross']=df['Worldwide Lifetime Gross'].map(lambda x: locale.atof(x.strip('$')))
```

```python
df['Domestic Lifetime Gross']=df['Domestic Lifetime Gross'].map(lambda x: locale.atof(x.strip('$')))
```

```python
df['Foreign Lifetime Gross']=df['Foreign Lifetime Gross'].map(lambda x: locale.atof(x.strip('$')))
```

```python
df['Rank']=df['Rank'].map(lambda x: int(x.replace(',','')))
```

```python
df['Rank']
```

```python
df['Domestic %']=df['Domestic %'].map(lambda x: x.replace('-','0%'))
df['Domestic %']
```

```python
df['Domestic %']=df['Domestic %'].map(lambda x: x.replace('<',''))
df['Domestic %']
```

```python
df['Domestic %']=df['Domestic %'].map(lambda x: float(x.replace('%','')))
df['Domestic %']
```

```python
df['Foreign %']=df['Domestic %'].map(lambda x: 100-x)
df['Foreign %']
```
