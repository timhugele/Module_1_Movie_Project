```python
'''Importing libraries'''
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

'''Assigning DataFrames'''
df_imdb=pd.read_csv('imdb_cleaned_data.csv')
df_mojo=pd.read_csv('box_office_mojo_cleaned.csv')
df_joined=pd.read_csv('merged_dataframes.csv')
```

```python
df_joined.head() #Checking headers
```

```python
df_joined=df_joined.drop('Unnamed: 0',axis=1)
df_joined.head() #Dropping unnecessary columns
```

```python
df_mojo=df_mojo.drop('Unnamed: 0',axis=1)
df_mojo.head()
```

```python
df_imdb=df_imdb.drop('Unnamed: 0',axis=1)
df_imdb.head()
```

```python
# Linear regression scatterplot using seaborn by year and domestic % of revenue
plt.figure(figsize=(12,8))
sns.regplot(df_mojo.Year,df_mojo['Domestic %'])
plt.title('% Domestic Lifetime Gross by Year')
plt.show()
```

```python
# Linear regression for Foreign % and year
plt.figure(figsize=(12,8))
sns.regplot(df_mojo.Year,df_mojo['Foreign %'])
plt.title('% Foreign Lifetime Gross by Year')
plt.show()
```

```python
# Heat map of Foreign % of revenue and year
sns.jointplot(df_mojo.Year,df_mojo['Foreign %'],kind='kde')
```

```python
# Domestic % and year heatmap
sns.jointplot(df_mojo.Year,df_mojo['Domestic %'],kind='kde')
```

```python
#correlation between Foreign % and Year
df_mojo.Year.corr(df_mojo['Foreign %'])
```

```python
#Correlation between Domestic % and Year
df_mojo.Year.corr(df_mojo['Domestic %'])
```

```python
#Heatmap of Box Office Rank by Year 
sns.jointplot(df_mojo.Year,df_mojo.Rank,kind='kde')
```

```python
#Dataframe of top 20 all time gross
df_20=df_mojo.head(20)
```

```python
df_20
```

```python
#horizontal bar chart of Top 20 Worldwide Gross
plt.figure(figsize=(10,8))
sns.barplot(df_20['Worldwide Lifetime Gross'],df_20.Title,palette='Blues_d')
plt.title('Top 20 Movies by Worldwide Lifetime Gross $bn')
```

```python
#Barchart of $ amount foreign and domestic of top 20 
df_20.plot(x='Title',y=["Domestic Lifetime Gross", "Foreign Lifetime Gross"], kind="bar")
plt.show()
```

```python
# making dataframe with Title as index
df_title_20 = df_20.set_index('Title')
```

```python
# making dataframe consist of only % Domestic and Foreign
df_title_20 = df_title_20[["Domestic %", "Foreign %"]]
```

```python
df_title_20.head()
```

```python
#Better bar chart of Domestic and Foreign %

#creating x axis 
x = np.arange(20)

#width to space bars
w = 0.3 
plt.figure(figsize=(12,8))

#plotting bar chart with tick marks placed between bars
plt.xticks(x+w/2, df_title_20.index, rotation = 'vertical')
ax=plt.subplot(111)

ax.bar(x, df_title_20['Domestic %'], width=w, label='Domestic %')

#spacing second bar so it doesn't cover up the first bar
ax.bar(x+w, df_title_20['Foreign %'], width=w,label='Foreign %')

plt.legend()
plt.xlabel('Title')
plt.ylabel('Percentage of Total Lifetime Gross')
plt.title('Top 20 Movies % Revenue Foreign and Domestic')
plt.show()
```

```python
# dataframe for $ amount of gross
df_gross_20 = df_20.set_index('Title')
df_gross_20 = df_gross_20[["Domestic Lifetime Gross", "Foreign Lifetime Gross"]]
df_gross_20.head()
```

```python
#Bar chart for top 20 movies 

#Copied from previous bar chart, changed variables to avoid confusion
x1 = np.arange(20)
w1 = 0.3
plt.figure(figsize=(12,8))

plt.xticks(x1+w1/2, df_gross_20.index, rotation = 'vertical')
ax=plt.subplot(111)

ax.bar(x1, df_gross_20['Domestic Lifetime Gross'], width=w, label='Domestic')
ax.bar(x1+w1, df_gross_20['Foreign Lifetime Gross'], width=w,label='Foreign')
plt.legend()
plt.xlabel('Movie Title')
plt.ylabel('Revenue in Billions USD')
plt.title('Top 20 Box Office Movies Foreign and Domestic Gross')
plt.show()
```

```python
# Creating dataframe of top 50 box office gross
df_50=df_mojo.head(50)
```

```python
# sum of top 50 gross revenue
dom_50_revenue=df_50['Domestic Lifetime Gross'].sum()
```

```python
# sum of foreign revenue top 50
for_50_revenue=df_50['Foreign Lifetime Gross'].sum()
```

```python
# sum of total worldwide revenue top 50
tot_50_revenue=df_50['Worldwide Lifetime Gross'].sum()
```

```python
print(dom_50_revenue/tot_50_revenue)
print(for_50_revenue/tot_50_revenue)
print(tot_50_revenue)
print(dom_50_revenue)
print(for_50_revenue)
```

```python
# Pie chart for percentage domestic and foreign for top 50
labels = ['Domestic', 'Foreign']
# sizes from percentages calculated above
sizes = [35.1, 64.8]
colors = ['gold', 'yellowgreen']

# Plot
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.title('Box Office Revenue of Top 50 Grossing Movies')
plt.axis('equal')
plt.show()
```
