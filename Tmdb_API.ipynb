```python
html_page = requests.get('https://www.imdb.com/search/title/?release_date=2020&sort=num_votes,desc&page=1')
soup = BeautifulSoup(html_page.content, 'html.parser')
```

```python
movie_titles = []
index_list = list(range(0,50))

for index in index_list:
    movie_titles.append(titles[index].find('a').string)

# movie_titles
```

```python
movie_info = movie_container.findAll('div', class_ = 'ratings-bar')
```

```python
movie_info[1].find('strong').text
```

```python
imdb_ratings = []
index_list2 = list(range(0,50))

for index2 in index_list2:
    imdb_ratings.append(movie_info[index2].find('strong').text)
    
# movie_ratings
```

```python
meta_scores = []
index_list3 = list(range(0,50))

for index3 in index_list3:
    meta_scores.append(movie_info[index3].find('span').nextSibling.nextSibling.text)

# metacritic_scores
```

```python
df2 = pd.DataFrame([movie_titles, imdb_ratings, meta_scores]).transpose()
df2.columns = ['Title', 'IMDb Rating', 'Metacritic Rating']
```
