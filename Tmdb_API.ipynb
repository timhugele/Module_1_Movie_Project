```python
imdb_ratings2 = []
page_range = list(range(1,2000,50))

for page in page_range:
    html_page2 = requests.get('https://www.imdb.com/search/title/?release_date=1990-01-01,2020-03-03&sort=boxoffice_gross_us,desc&start={}&ref_=adv_nxt'.format(page))
    soup2 = BeautifulSoup(html_page2.content, 'html.parser')
    movie_container5 = soup2.find('div', class_="lister list detail sub-list")
    movie_info = movie_container5.findAll('div', class_ = 'ratings-bar')
    
    imdb_r_index = list(range(0,50))
    for index2 in imdb_r_index:
        if movie_info[index2].find('strong'):
            imdb_ratings2.append(float(movie_info[index2].find('strong').text))
        else:
            imdb_ratings2.append('None')
        
    
imdb_ratings2[:5]
```

```python
# THIS CELL PRODUCES THE METACRITIC RATINGS
```

```python
metac_ratings = []
page_range = list(range(1,2000,50))

for page in page_range:
    html_page2 = requests.get('https://www.imdb.com/search/title/?release_date=1990-01-01,2020-03-03&sort=boxoffice_gross_us,desc&start={}&ref_=adv_nxt'.format(page))
    soup2 = BeautifulSoup(html_page2.content, 'html.parser')
    movie_container5 = soup2.find('div', class_="lister list detail sub-list")
    movie_info = movie_container5.findAll('div', class_ = 'ratings-bar')
    
    metac_index = list(range(0,50))
    for index in metac_index:
        if movie_info[index].find('span', class_="metascore favorable"):
            metac_ratings.append(float(movie_info[index].find('span', class_="metascore favorable").text.strip()))
        elif movie_info[index].find('span', class_="metascore mixed"):
            metac_ratings.append(float(movie_info[index].find('span', class_="metascore mixed").text.strip()))
        elif movie_info[index].find('span', class_="metascore unfavorable"):
            metac_ratings.append(float(movie_info[index].find('span', class_="metascore unfavorable").text.strip()))
        else:
            metac_ratings.append('None')
        
    
metac_ratings[:5]
```

```python
# THIS CELL PRODUCES THE YEAR THAT THE MOVIE WAS RELEASED
```

```python
movie_years = []
page_range = list(range(1,2000,50))

for page in page_range:
    html_page = requests.get('https://www.imdb.com/search/title/?release_date=1990-01-01,2020-03-03&sort=boxoffice_gross_us,desc&start={}&ref_=adv_nxt'.format(page))
    soup = BeautifulSoup(html_page.content, 'html.parser')
    movie_container = soup.find('div', class_="lister list detail sub-list")
    movie_info = movie_container.findAll('span', class_ = 'lister-item-year text-muted unbold')

    year_index = list(range(0,50))
    for index in year_index:
        if movie_info[index].text.startswith("(I)"):
            year = movie_info[index].text[5:]
            year = year[:-1]
            movie_years.append(year)
        else:
            year = movie_info[index].text[1:]
            year = year[:-1]
            movie_years.append(year)

movie_years[:5]
```
